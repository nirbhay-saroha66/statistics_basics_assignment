**Q1)Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.**

Ans) Data can be classified into two types: qualitative and quantitative.

1.**Qualitative Data (Categorical Data)**

**Definition:** Represents characteristics or qualities that are descriptive in nature and cannot be measured numerically.

**Types:**

a) **Nominal Scale:** Data with categories that have no inherent order or ranking.

Examples:
Eye color (blue, green, brown).
Marital status (single, married, divorced).

b) **Ordinal Scale:** Data with categories that have a meaningful order or ranking, but the differences between ranks are not uniform or meaningful.

Examples:
Education level (high school, undergraduate, graduate).
Likert scale responses (strongly agree, agree, neutral, disagree, strongly disagree).

**2. Quantitative Data (Numerical Data):**

**Definition:** Represents measurable quantities and is expressed in numbers.

**Types:**

a) **Interval Scale:** Numeric data where the differences between values are meaningful, but there is no true zero point.

Examples:
Temperature in Celsius or Fahrenheit (0Â°C does not mean "no temperature").
Dates (e.g., 2025, 1990).

**Ratio Scale:** Numeric data with a true zero point, meaning the absence of the quantity being measured. Ratios are meaningful.

Examples:
Height (e.g., 180 cm).
Weight (e.g., 70 kg).
Income (e.g., $50,000).

**Q2)What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate**

ans)
1. **Mean**

**Definition**: The arithmetic average of all values in a dataset.

**Mean** = (Sum of all values/Number of values)

**Example:**

**Dataset:** 5,10,15,20,25

**Mean:** (5+10+15+20+25)/5 =15

**When to Use:**

Use when the data is quantitative and the distribution is symmetric without outliers.

Example: Average test scores of students.

**Not Suitable When**:

The data contains outliers that can skew the mean (e.g., income data).

2. **Median**

**Definition**: The middle value when the data is arranged in ascending or descending order. For an even number of values, the median is the average of the two middle values.

Example:

Dataset: (5,10,15,20,25)

Median = 15

Dataset: (5,10,15,20)

Median: (10+15)/2 = 12.5

**When to Use:**

Use for quantitative data with skewed distributions or outliers.

Example: Median income of a population when a few individuals earn exceptionally high amounts.

Advantage: Not affected by extreme values.

3. **Mode**

**Definition:** The value(s) that appear most frequently in the dataset. A dataset can have one mode (unimodal), more than one mode (bimodal, multimodal), or no mode.

Example:

Dataset: (5,10,10,15,20)

mode: 10

**When to Use:**

Use for categorical or discrete data to identify the most common category or value.

Example: The most popular ice cream flavor in a survey.

**Q3) Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**

ans) Dispersion
Dispersion measures the spread or variability of data. It indicates how much data points differ from each other and from the mean.

Variance
Definition: The average of the squared differences from the mean.

Purpose: Quantifies how spread out data is. Larger variance means greater spread.

Standard Deviation

Definition: The square root of variance, giving spread in the same units as the data.

Purpose: Easier to interpret than variance.


**Q4) What is a box plot, and what can it tell you about the distribution of data?**

ans) **Box Plot**

A box plot (or whisker plot) is a graphical representation of the distribution of a dataset. It summarizes key statistics in a compact visual form, showing the spread and identifying potential outliers.

**Components of a Box Plot**

1.Median (line inside the box): Divides the data into two halves.

2.Box:

a)Lower Quartile (Q1): 25th percentile (bottom edge of the box).

b)Upper Quartile (Q3): 75th percentile (top edge of the box).

c)Interquartile Range (IQR): Q3-Q1, showing the middle 50% of the data.

3.Whiskers:

Extend from Q1 and Q3 to the smallest and largest values within
1.5xIQR.

Outliers: Data points beyond 1.5xIQR from Q1 or Q3, shown as dots or asterisks.

**What It Tells You About Data**

1.Spread:

a)The length of the box shows the range of the middle 50% of data (IQR).

b)Whiskers indicate variability in the data outside the IQR.

2.Central Tendency:

a)The median line indicates the central value.

3.Symmetry or Skewness:

a)A symmetrical box suggests a roughly even distribution.

b)A shifted median or uneven whiskers indicate skewness.

4.Outliers:

a)Points outside the whiskers highlight potential anomalies.


**Q5)  Discuss the role of random sampling in making inferences about populations**

**Role of Random Sampling in Making Inferences:**

Random sampling is a fundamental technique in statistics used to make inferences about a population based on data collected from a subset (sample). It ensures that every individual in the population has an equal chance of being selected, making the sample representative of the population.

**Key Benefits of Random Sampling**

1.Minimizes Bias:

a)Ensures no systematic favoritism in sample selection.

b)Results in a sample that mirrors the population's diversity.

2.Supports Generalization:

Findings from the sample can be generalized to the entire population with a degree of confidence.

3.Facilitates Statistical Analysis:

Allows the use of probability theory to compute estimates (e.g., mean, proportion) and their associated margins of error.

4.Provides Accuracy and Precision:

Larger random samples reduce variability and increase the reliability of estimates.

**Limitations**

1.Sampling Errors:

Natural variations due to chance in sample selection.

2.Non-Response Bias:

When certain individuals do not respond, it can distort results.

3.Practical Challenges:

Ensuring randomness in large, complex populations can be difficult.

**Q6) Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**

ans) Skewness is a measure of the asymmetry in the distribution of data around its mean. It indicates whether data points are concentrated more on one side of the distribution or evenly distributed.

Symmetric Distribution: Skewness = 0 (e.g., normal distribution).

Asymmetric Distribution: Skewness â‰  0.

**Types of Skewness**

1.**Positive Skew (Right-Skewed):**

a)The tail is longer on the right side of the distribution.

b)Most data points are concentrated on the left.

c)Examples: Income distribution, housing prices.

d)Effect: Mean > Median > Mode.

2.**Negative Skew (Left-Skewed):**

a)The tail is longer on the left side of the distribution.

b)Most data points are concentrated on the right.

c)Examples: Age of retirement, exam scores with many high achievers.

d)Effect: Mean < Median < Mode.

**3.Symmetric:**

a)Equal tails on both sides.

b)Examples: Heights of individuals (approximately normal).

c)Effect: Mean â‰ˆ Median â‰ˆ Mode.

**Effects of Skewness on Data Interpretation**

**1.Central Tendency:**

a)Skewness affects the relationship between mean, median, and mode.

b)Mean is pulled in the direction of the tail, making it sensitive to outliers.

**2.Outlier Detection:**

Skewness highlights the presence of extreme values (long tails).

**3.Choice of Statistical Measures:**

a)For skewed data, the median is a better measure of central tendency than the mean.

b)Use non-parametric tests for highly skewed data.

**4.Visual Impact:**

Skewed data can mislead interpretations if only summary statistics (e.g., mean) are considered without visualization.

**Q7)What is the interquartile range (IQR), and how is it used to detect outliers?**

ans) Interquartile Range (IQR)
The interquartile range (IQR) is a measure of statistical dispersion that represents the range within which the middle 50% of the data lies. It is calculated as the difference between the third quartile and first quartile

IQR = Q3 - Q1

Q1 (First Quartile): The median of the lower half of the data (25th percentile).

Q3 (Third Quartile): The median of the upper half of the data (75th percentile).

IQR: Represents the spread of the middle 50% of the dataset.

**Using IQR to Detect Outliers**

Outliers are data points that fall significantly outside the typical range of a dataset. The IQR is used to define thresholds for detecting these outliers:

1.Calculate the IQR:

Find Q1 and Q3.

Compute IQR=Q3-Q1.

2.Determine Outlier Boundaries:

Lower Boundary: Q1-1.5xIQR.

Upper Boundary: Q3+1.5xIQR.

Identify Outliers:

Any data point below the lower boundary or above the upper boundary is considered an outlier.

Why IQR for Outlier Detection?

Robustness: Unlike the mean, the IQR is not affected by extreme values, making it ideal for detecting outliers in skewed or non-normal distributions.

Applications: Used in box plots and to clean data in fields like finance, healthcare, and social sciences.

**Q11) Explain what a random variable is and differentiate between discrete and continuous random variables.**

ans)A random variable is a numerical representation of the outcomes of a random phenomenon. It assigns a value to each possible outcome of a random experiment.

Notation: Random variables are typically denoted by uppercase letters (e.g.,
ð‘‹
X).
Types of Random Variables: Discrete and Continuous.

Discrete Random Variables

Definition:

A random variable that takes on a countable number of distinct values.

Key Features:

Values are often whole numbers.
Each value has a specific probability.
Examples:

Number of heads in 10 coin tosses.
Number of customers arriving at a store in an hour.
Probability Distribution:

Represented by a probability mass function (PMF), which gives the probability for each value of the variable.

Continuous Random Variables

Definition:
A random variable that can take on an infinite number of possible values within a range.

Key Features:

Values are uncountable (e.g., real numbers).
Probabilities are assigned to intervals, not specific values.
Examples:

Time taken to complete a task.
Height or weight of individuals.
Probability Distribution:

Represented by a probability density function (PDF), where the area under the curve over an interval represents the probability.
